```python
import json
import os
import numpy as np
import cv2      
from shapely.geometry import Polygon
import numpy as np
import time 
from shapely.validation import make_valid
import warnings
import matplotlib.pyplot as plt

warnings.filterwarnings(action='ignore')

    
def load_json(json_file) : 

    print(f'START {json_file}')
    with open(json_file, 'r') as file:
        original_json_dict = json.load(file)
        total_points_list = []
        for seg_idx, j_dict in enumerate(original_json_dict['seg']) :
            seg_points = j_dict['points']
            points_dict = {}
            points_list = []

            for j_point in seg_points :
                x_y = [j_point['x'],j_point['y']]
                points_list.append(x_y)



            points_list = np.array(points_list)
            points_list = points_list.reshape((-1, 2))
            pts = np.array(points_list,np.int32)
            pts = pts.reshape((-1, 1, 2))

            total_points_list.append(points_list)

    return total_points_list

def draw_line_img(total_points_list,img_path):
    image_ori = cv2.imread(img_path)
    for polygon in total_points_list :
        pts = np.array(polygon,np.int32)
        pts = pts.reshape((-1, 1, 2))

        color_np = np.asarray(np.random.choice(range(256), size=3), dtype=np.uint8)
        color_fill = (color_np[0].item(),color_np[1].item(),color_np[2].item())

        isClosed = True
        color = (255, 0, 0)
        thickness = 2
        line_image = cv2.polylines(image_ori, [polygon], isClosed, color_fill, thickness)

    f = plt.figure()
    f.set_figwidth(20)
    f.set_figheight(15)
    plt.imshow(line_image)
    plt.show()

def draw_fill_poly_img(total_points_list,img_path):
    image_ori = cv2.imread(img_path)
    for polygon in total_points_list :
#         pts = np.array(polygon,np.int32)
#         pts = pts.reshape((-1, 1, 2))


        color_np = np.asarray(np.random.choice(range(256), size=3), dtype=np.uint8)
        color_fill = (color_np[0].item(),color_np[1].item(),color_np[2].item())
        

        fill_image_ori = cv2.fillPoly(image_ori, pts=[polygon], color=color_fill)

    f = plt.figure()
    f.set_figwidth(20)
    f.set_figheight(15)
    plt.imshow(fill_image_ori)
    plt.show()
        
json_file = 'SC_PAR_20220824_185745_cam01.json'   
json_path = 'ij/SC_PAR_20220824_185745_cam01.json' 
img_file = json_file[:-4] + 'jpg'
img_path = 'aihub_img/' + img_file

total_points_list = load_json(json_path)
draw_line_img(total_points_list,img_path)
draw_fill_poly_img(total_points_list,img_path)
```
